```python
from pathlib import Path
data_folder = Path("../Dataset")
output_folder = Path("../Dataset_rev")

file_users = data_folder / "users.csv"
output = output_folder / "users_rev.csv"
```

```python
import pandas as pd

df_users = pd.read_csv(file_users)
df_users
```

## cordinates -> latitude, longitude

```python
coordinates = [eval(coord) for coord in df_users['coordinates']]

latitude, longitude = zip(*coordinates)

df_users['user_latitude'] = latitude
df_users['user_longitude'] = longitude

df_users.drop(columns=['Unnamed: 0', 'coordinates'], inplace=True)
df_users.rename(columns={'created_at':'user_created_at'}, inplace=True)
```

```python
df_users
```

## Get location from latitude and longitude

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="user_location")

city = []
state = []
country = []

for coord in coordinates:
    location = geolocator.reverse(coord, language="en")
    
    try:
        address = location.raw['address']

        # traverse the data
        city.append(address.get('city', ''))
        state.append(address.get('state', ''))
        country.append(address.get('country', ''))
        
    except:
        city.append('')
        state.append('')
        country.append('')
```

```python
df_users['user_city'] = city
df_users['user_state'] = state
df_users['user_country'] = country
```

```python
df_users
```

```python
df_users.to_csv(output)
```
