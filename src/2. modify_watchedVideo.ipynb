```python
from pathlib import Path
data_folder = Path("../Dataset")
output_folder = Path("../Dataset_rev")

file_users = data_folder / "watchedVideo.csv"
output = output_folder / "watchedVideo_rev.csv"
```

```python
import pandas as pd

df_watchedVideo = pd.read_csv(file_users)
df_watchedVideo
```

## cordinates -> latitude, longitude

```python
import re

coordinates = [eval(re.sub('[^0-9,.-]', '',df_watchedVideo.iloc[i]['coordinates'])) for i in range(50000)]

latitude, longitude =  zip(*coordinates)

df_watchedVideo['latitude'] = latitude
df_watchedVideo['longitude'] = longitude


df_watchedVideo.drop(columns=['Unnamed: 0', 'coordinates'], inplace=True)
```

```python
df_watchedVideo
```

## Get location from latitude and longitude

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="watched_location")

def getCoordinates(part_coordinates, city, state, country):
    for coord in part_coordinates:
        try:
            location = geolocator.reverse(coord, language="en")        
            address = location.raw['address']

            # traverse the data
            city.append(address.get('city', ''))
            state.append(address.get('state', ''))
            country.append(address.get('country', ''))

        except:
            city.append('')
            state.append('')
            country.append('')
            
    return city, state, country
        
```

```python
city = []
state = []
country = []
```

```python
import time
for i in range(0, len(coordinates), 1000):
    city, state, country = getCoordinates(coordinates[i:i+1000], city, state, country)
    time.sleep(1)

    
```

```python
df_watchedVideo['user_city'] = city
df_watchedVideo['user_state'] = state
df_watchedVideo['user_country'] = country
```
