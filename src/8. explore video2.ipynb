```python
from pathlib import Path

data_folder = Path("../Dataset_rev")
file = data_folder / "videos_rev2.csv"
```

```python
import pandas as pd

df = pd.read_csv(file)
df['Month']= pd.to_datetime(df['user_created_at']).dt.month
df['numWatched'] = df['numWatched'].fillna(0)
df
```

```python
# correlation
df_corr = df.drop(columns = ['id','user_latitude', 'user_longitude']).apply(lambda x: x.factorize()[0]).corr()
df_corr
```

```python
corr_set = set()
for col in df_corr:
    for index in df_corr.index:
        if df_corr.loc[index].loc[col] >= 0.5 and index != col:
            corr_set.add(tuple(sorted((index, col))))
```

```python
def getCount_col (col):
    cnt_df = df.groupby(df[col]).size().reset_index(name='counts')
    cnt_df.set_index(col, inplace=True)
    cnt_df.sort_values(by=['counts'], ascending=False, inplace=True)
    
    return cnt_df
```
